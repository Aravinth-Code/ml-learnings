```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("E:\Projects\ml-learnings\Project Dependencies\homeprices.csv")
df
```

```python
plt.xlabel("area(sq.ft)")
plt.ylabel("price(USD)")
plt.scatter(df.area, df.price, color = 'red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3300]])
```

```python
dataframe = pd.read_csv("E:\Projects\ml-learnings\Project Dependencies\hareas.csv")
```

```python
predicted_dataframe = reg.predict(dataframe)
predicted_dataframe
```

```python
dataframe['prices'] = predicted_dataframe
```

```python
dataframe.to_csv("E:\Projects\ml-learnings\Project Dependencies\predicted_areas.csv", index=False)
```
