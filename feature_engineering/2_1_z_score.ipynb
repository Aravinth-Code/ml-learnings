```python
import pandas as pd
df = pd.read_csv("E:\Projects\ml-learnings\project_dependencies\weight-height.csv")
df.head()
```

```python
df.describe()
```

```python
import matplotlib as mb
import matplotlib.pyplot as plt
plt.hist(df.Height, bins=20, rwidth=0.8)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
from scipy.stats import norm
import numpy as np
plt.hist(df.Height, bins=20, rwidth=0.8)
plt.xlabel("Height")
plt.ylabel("Weight")
rng = np.arange(df.Height.min(), df.Height.max(), 0.2)
plt.plot(rng, norm.pdf(rng, df.Height.mean(), df.Height.std()))
```
